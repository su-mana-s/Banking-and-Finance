```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
historic = pd.read_csv("../../data/interim/UHistoric_Domestic.csv")
```

```python
date = historic["date"]
```

```python
rgdp = historic["real gdp growth"].to_list()
fig, ax = plt.subplots()
ax.plot(rgdp)
labels = {}
for i in range(len(date)):
    labels[i] = date[i]

def format_fn(tick_val, tick_pos):
    if int(tick_val) >= 0 and int(tick_val) < 94:
        return labels[int(tick_val)]
    else:
        return ''
ax.xaxis.set_major_formatter(format_fn)

plt.show()
```

```python
rgdp = historic["real gdp growth"].to_list()
ur = historic["unemployment rate"].to_list()
nrgdp = historic["nominal gdp growth"].to_list()
plt.plot(rgdp)
plt.plot(nrgdp)
x = [i for i in range(0, len(date), 25)]
labels = [date[i] for i in x]
plt.xticks(x, labels, rotation ='vertical') 
plt.show()
```
