```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
liq = pd.read_csv("../../outputs/liquidity.csv")
liq.head()
```

```python
q = liq["quickRaito"][1:]
c=  liq["currentRatio"][1:]
qa = liq["quickAssets"][1:]	
ca = liq["currentAssets"][1:]	
cl = liq["currentLiabilities"][1:]	
rats = {'qa': qa, 'ca':ca, 'cl':cl}
labels = liq["Date"][1:]
```

What is my y axis even doing?

```python
x = np.arange(4)  # the label locations

width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in rats.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    # ax.bar_label(measurement, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
# ax.set_ylabel('Ratio')
# ax.set_title('Quick and Current Ratios')
ax.set_xticks(x + width, labels)
ax.legend(loc='upper left')
# ax.set_ylim(-0, 8)
plt.show()
```
