US Banks Basic Normalization

```python
import boto3
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# AWS credentials and region
aws_access_key_id = 'AKIAW3MED5OHMTBP3VFW'
aws_secret_access_key = '/UG/lOcC9fY3PbBhSxSecPnLX4i6V1kLR29TUZd8'
region_name = 'ap-south-1'  # e.g., 'us-east-1'
```

```python
# Initialize S3 client
s3 = boto3.client('s3', 
                  aws_access_key_id=aws_access_key_id,
                  aws_secret_access_key=aws_secret_access_key,
                  region_name=region_name)
```

```python
# S3 bucket and file key
bucket_name = 'usbanksbasic'
file_key = 'Final_USBankBasics.csv'
```

```python
# Download the file from S3 and load into DataFrame
response = s3.get_object(Bucket=bucket_name, Key=file_key)
data = pd.read_csv(response['Body'],header=0)
data.head()
```

```python
first_two_columns = data[['Tag', 'Bank']]
first_two_columns.head()
```

```python
# Exclude the columns "Breakdown" and "Bank" from the DataFrame
data_subset = data.drop(columns=['Tag', 'Bank'])
```

```python
# Perform data normalization (example: Min-Max scaling) on remaining columns
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(data_subset)
```

```python
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(data.iloc[:, 2:])
```

```python
# Convert normalized data back to DataFrame
normalized_df = pd.DataFrame(normalized_data, columns=data_subset.columns)
```

```python
# Concatenate the columns "Breakdown" and "Bank" with the normalized DataFrame
normalized_df_with_columns = pd.concat([first_two_columns, normalized_df], axis=1)
normalized_df_with_columns.head()
```
